```python
import numpy as np
import RBF
import matplotlib.pyplot as plt
```

```python
x = (np.arange(2*np.pi *10)/10) [:,np.newaxis]   ##generate data [0,2*pi], interval is 0.1
y1 = np.sin(2*x)
y2 = np.square(2*x)
```

```python
x.shape
```

```python
x_test= (np.arange(0.5,2*np.pi*10)/10)[:,np.newaxis]
y1_test = np.sin(2*x_test)
y2_test = np.square(2*x_test)
nodes_list=[]
errors_list=[]
models_list=[]
```

```python
for nodes_num in range(10):
    model = RBF.RBF(sigma=0.001, hidden_layer_nodes=nodes_num)
    model.fit_least_squares(x, y1)
    y_pred = model.predict(x_test)
    errors = np.sum(np.abs(y_pred-y1_test))/len(y_pred)
    if errors <0.1:
        nodes_list.append(nodes_num)
        errors_list.append(errors)
        models_list.append(model)
    elif errors <0.01:
        nodes_list.append(nodes_num)
        errors_list.append(errors)
        models_list.append(model)
    elif errors <0.001:
        nodes_list.append(nodes_num)
        errors_list.append(errors)
        models_list.append(model)
        break 
```

```python
errors_list
```

```python
model = RBF.RBF(sigma=0.001, hidden_layer_nodes=10)
model.fit_least_squares(x, y1)
y_pred = model.predict(x_test)
plt.plot(x_test, y1_test, 'b-', label='real')
plt.plot(x_test, y_pred, 'r-', label='fit')
plt.legend(loc='upper right')
plt.title('Interpolation using a RBFN')
plt.show()
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv("ball/balllist.csv")[0:100]
dfx=pd.DataFrame(df,columns=['x1','x2'])
dfy=pd.DataFrame(df,columns=['y1','y2'])
x_train = dfx.values
y_train = dfy.values

df=pd.read_csv("ball/balltest.csv")[0:100]
dfx=pd.DataFrame(df,columns=['x1','x2'])
dfy=pd.DataFrame(df,columns=['y1','y2'])
x_test = dfx.values
y_test = dfy.values
```

```python
import RBF_with_competitivelearning as RBF_CL
```

```python
errors=[]
for node in range(1,170):
    model = RBF_CL.RBF_with_competitivelearning(sigma=0.1, hidden_layer_nodes=node,CL_epoch=10000,eta_CL=0.01)
    model.fit_least_squares(x_train, y_train)
    y_pred = model.predict(x_test)
    errors.append(np.sum(np.sqrt(np.sum((y_pred-y_test)*(y_pred-y_test),axis=1)))/100)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,170), errors, 'r-', label='fit')
plt.legend(loc='upper right')
plt.title('loss-nodes')
plt.xlabel("number of nodes")
plt.ylabel("loss")
plt.show()
```
